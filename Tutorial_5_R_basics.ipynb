## 1. Variables and Data Types

Unlike some other languages that use `=` for assignment, R traditionally uses the leftward arrow `<-` to assign values to variables. While `=` does work, `<-` is the standard practice in the R community.

R has a few basic data types you'll use constantly:
* **Numeric:** Decimal values (e.g., 10.5, 4)
* **Character:** Text or string values (e.g., "Hello")
* **Logical:** Boolean values (`TRUE` or `FALSE`)

## 2. Vectors

Vectors are the most basic data structure in R. A vector is simply a list of items that are of the **same data type**.

To create a vector, we use the `c()` function, which stands for **combine** or **concatenate**. R is highly optimized for vector operations, meaning you can often perform math on a whole list of numbers at once without needing a loop.

```python
%%R

scores <- c(85, 92, 78, 90, 88)

print(scores)

bonus_scores <- scores + 5
print(bonus_scores)

print(mean(scores)) # Calculates the average
print(max(scores))
```

## 3. Data Frames

In data science, we rarely work with single vectors. Usually, we work with datasets that look like spreadsheets. In R, this is called a **Data Frame**.

A Data Frame is essentially a collection of vectors of the same length, where each vector is a column.

```python
%%R
students_df <- data.frame(
  Name = c("Rahul", "Sneha", "Amit"),
  Age = c(20, 21, 19),
  Passed = c(TRUE, TRUE, FALSE)
)

print(students_df)

print(students_df$Name)
```

## 4. Lists (Mixing Data Types)

Earlier, we learned that **Vectors** can only hold one type of data (all numbers, or all text). But what if you need to store different types of data together? That's where **Lists** come in.

A list is a collection that can hold numbers, strings, vectors, and even other lists!

```python
%%R
# Creating a list with different types of data
student_profile <- list(
  Name = "Priya",
  Age = 22,
  Scores = c(95, 88, 92),
  Is_Enrolled = TRUE
)

# Printing the entire list
print(student_profile)

# Accessing a specific item in the list using the $ operator
print(student_profile$Scores)
```

## 5. Control Flow: If-Else Statements

Sometimes, we only want our code to run *if* a certain condition is met. We use `if` and `else` statements to make decisions in our code.

```python
%%R
# Define a test score
score <- 75

# Check if the student passed or failed
if (score >= 50) {
  print("Congratulations, you passed!")
} else {
  print("You did not pass. Try again next time.")
}

# You can also chain conditions using 'else if'
temperature <- 30

if (temperature > 35) {
  print("It's a hot day!")
} else if (temperature >= 20) {
  print("The weather is pleasant.")
} else {
  print("It's a bit cold today.")
}
```

## 6. Loops (Repeating Actions)

If you want to perform the same action multiple times, you don't need to copy and paste your code. You can use a **loop**. The most common loop is the `for` loop, which iterates over a sequence of items.

```python
%%R
# A simple loop to print numbers 1 through 5
for (i in 1:5) {
  print(paste("This is loop iteration number:", i))
}

# Looping through a vector of names
fruits <- c("Apple", "Banana", "Mango", "Grapes")

for (fruit in fruits) {
  print(paste("I like eating", fruit))
}
```

## 7. Creating Your Own Functions

While R has many built-in functions like `mean()` or `max()`, you can also write your own! Functions are great for wrapping up a block of code so you can reuse it easily without rewriting it.

```python
%%R
# Define a custom function to calculate the square of a number
calculate_square <- function(number) {
  result <- number * number
  return(result)
}

# Use (call) the custom function
print(calculate_square(4))
print(calculate_square(10))

# A function that takes two arguments
add_numbers <- function(num1, num2) {
  return(num1 + num2)
}

print(add_numbers(15, 25))
```

## 8. Handling Missing Data (NA)

In the real world, data is rarely perfect. You will often have missing values. In R, missing data is represented by `NA` (Not Available).

It is crucial to know how to find and handle `NA` values because they can break your calculations!

```python
%%R
# A vector with a missing value
sales <- c(150, 200, NA, 300, 250)
print(sales)

# Check which values are missing (returns TRUE for NA)
print(is.na(sales))

# Trying to calculate the mean with an NA value returns NA
print(mean(sales))

# To fix this, we tell R to ignore (remove) the NA values: na.rm = TRUE
print(mean(sales, na.rm = TRUE))
```

## 9. Factors (Categorical Data)

When working with data, you often have text that represents categories rather than just free-form words (e.g., "Low", "Medium", "High" or "Male", "Female"). In R, we use **Factors** to handle categorical data efficiently.

Factors are super important when you start doing machine learning or statistical modeling in R!

```python
%%R
# A standard character vector of t-shirt sizes
sizes_vector <- c("Small", "Large", "Medium", "Small", "Large")
print(sizes_vector)

# Convert the character vector into a Factor
sizes_factor <- factor(sizes_vector)

# Notice how R now identifies the distinct "Levels" (categories)
print(sizes_factor)

# We can even check what those levels are directly
print(levels(sizes_factor))
```

## 10. Reading and Writing Data (CSV Files)

Most of the time, you won't be typing your data directly into R. You will load it from a file, analyze it, and maybe save the results to a new file. The most common file type for this is a CSV (Comma Separated Values) file.

```python
%%R
# Let's recreate our simple data frame
my_data <- data.frame(
  ID = c(1, 2, 3),
  Score = c(88, 95, 72)
)

# 1. Writing data to a CSV file in your current folder
write.csv(my_data, file = "student_scores.csv", row.names = FALSE)
print("Data saved to student_scores.csv!")

# 2. Reading data back from a CSV file
imported_data <- read.csv("student_scores.csv")
print(imported_data)
```

## 11. Built-in Datasets

R comes with several built-in datasets, which is fantastic for learning and practicing. One of the most famous datasets in data science is the `iris` dataset, which contains measurements of different iris flowers.

Let's load it and take a look at how to quickly explore a dataset.

```python
%%R

# Load the dataset
data(iris)

# head() shows the first 6 rows of the dataset
head(iris)

# summary() provides quick descriptive statistics for every column!
summary(iris)

# dim() tells us the dimensions (rows and columns)
dim(iris)
```

## 12. Subsetting Data (Filtering)

Often, you don't want to look at the whole dataset; you only want specific rows or columns. In R, we can use square brackets `[rows, columns]` to slice our data, or use the `subset()` function for easier filtering.

```python
%%R
# 1. Using brackets: Get the first 3 rows and all columns
print(iris[1:3, ])

# 2. Using brackets: Get the first 5 rows of just the 1st and 2nd columns
print(iris[1:5, c(1, 2)])

# 3. Using subset(): Get only the flowers where Sepal.Length is greater than 7
large_flowers <- subset(iris, Sepal.Length > 7)
print(head(large_flowers))
```

## 13. Basic Data Visualization

R was built for statistics, so creating plots is incredibly straightforward. You don't even need to install extra packages to create scatter plots, histograms, or boxplots—Base R can do it right out of the box!

```python
%%R

# Histogram: Great for looking at the distribution of a single numeric variable
hist(iris$Sepal.Length,
     col = "lightblue",
     main = "Distribution of Sepal Lengths",
     xlab = "Sepal Length")

```

```python
%%R

# Scatter Plot: Great for comparing two numeric variables
plot(x = iris$Sepal.Length, y = iris$Sepal.Width,
     col = "darkred",
     main = "Sepal Length vs. Sepal Width",
     xlab = "Length", ylab = "Width",
     pch = 16) # pch = 16 makes the points solid circles
```
